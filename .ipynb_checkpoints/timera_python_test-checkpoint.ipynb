## Task
The file "day_ahead_power_prices.csv" contains hourly day ahead power prices in €/MWh for seven key European power markets from 2022-2024:
Belgium (BE), Germany (DE), Spain (ES), France (FR), Netherlands (NL), Poland (PL), and Great Britain (GB).

Hourly day-ahead power prices refer to the price of electricity agreed upon one day before delivery, set for each individual hour of the following day.

## Question 1

1) Calculate the monthly average prices for each country
2) The average of the highest 10%, 50% and 80% of prices for each country

```python
import pandas as pd
```

```python
file_path = 'day_ahead_power_prices.csv'
df = pd.read_csv(file_path)
df['DateTime'] = pd.to_datetime(df['DateTime'], dayfirst=True, errors='coerce')
df.set_index('DateTime', inplace=True)
```

```python
country_columns = [col for col in df.columns if col != 'DateTime']
monthly_avg = df[country_columns].resample('ME').mean()
monthly_avg.to_csv('monthly_average_prices.csv')
```

```python
results = []
for country in country_columns:
    prices = df[country].dropna()
    n = len(prices)
    results.append({
        'Country': country,
        'Top 10% Avg': prices.nlargest(int(n * 0.10)).mean(),
        'Top 50% Avg': prices.nlargest(int(n * 0.50)).mean(),
        'Top 80% Avg': prices.nlargest(int(n * 0.80)).mean(),
        'Missing Values': df[country].isna().sum()
    })
```

```python
percentile_df = pd.DataFrame(results)
percentile_df.to_csv('top_percentile_averages.csv', index=False)

print("Monthly Average Prices:\n", monthly_avg)
print("\nTop Percentile Averages:\n", percentile_df)
```

## Summary of Results – Question 1

### Monthly Average Prices
- France had the highest monthly average prices overall, followed by Germany.
- Spain consistently showed the lowest average prices across the period.
- All countries had complete data with no missing values, allowing reliable monthly averages apart from Great Britain with 4 missing value cells out of 7249 cells in total.

### Top Percentile Price Averages
- France also had the highest average among the top 10% of hourly prices (c. 453 €/MWh), indicating frequent extreme price spikes.
- Spain's top 10% average (c. 244 €/MWh) was significantly lower than the other countries.
- Across all countries, top 80% price averages were closer to each other, suggesting common pricing levels outside of peak extremes.

These results highlight France and Germany as high-price, high-peak markets, while Spain appears more stable and consistently lower-priced.

### Exploratory Analysis of Day-Ahead Power Prices (2022–2024)

This section presents four key visual analyses to explore the hourly day-ahead electricity prices across major European markets. The aim is to uncover trends, volatility, inter-market relationships, and price distribution features of interest to energy asset investors.

- **Trend Analysis:** Line plot of monthly average prices per country.
- **Volatility Analysis:** Box plot of price distribution per country.
- **Market Correlation:** Heatmap of price correlations between countries.
- **Price Distribution:** Kernel density estimates (KDE) for each country's price distribution.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 1. Line Plot: Monthly Average Prices
plt.figure(figsize=(14, 6))
for country in country_columns:
    plt.plot(monthly_avg.index, monthly_avg[country], label=country)
plt.title('Monthly Average Day-Ahead Power Prices (€/MWh)')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Box Plot: Distribution per Country
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[country_columns], showfliers=False)
plt.title('Price Distribution per Country')
plt.ylabel('€/MWh')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 3. Correlation Heatmap Between Countries
plt.figure(figsize=(8, 6))
corr_matrix = df[country_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Countries')
plt.tight_layout()
plt.show()
```

```python
# 4. KDE Plot: Price Distribution per Country
plt.figure(figsize=(14, 6))
for country in country_columns:
    sns.kdeplot(df[country].dropna(), label=country, linewidth=1.5)
plt.title('Price Distribution (Density Plot)')
plt.xlabel('€/MWh')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
skew_kurtosis = pd.DataFrame({
    'Skewness': df[country_columns].skew(),
    'Kurtosis': df[country_columns].kurtosis()
})
print(skew_kurtosis)
```

## Summary of Price Behavior Across European Power Markets (2022–2024)

This analysis explores hourly day-ahead power prices across key European countries, using monthly averages, price distributions, correlation, and density plots to highlight trends and characteristics relevant to investors.

### 1. Price Trends
- France and Germany show the highest average prices, especially during 2022.
- All countries trend downward through 2023–2024, with Spain remaining the most stable.

### 2. Volatility & Distribution
- France has the highest skew and kurtosis combination, indicating frequent extreme prices.
- Spain and Belgium have tighter, more symmetric price distributions, thus more stable.

### 3. Market Correlation
- Belgium, Netherlands, and Germany are highly correlated (>0.94), reflecting market coupling.
- Spain is less correlated, showing more independent price behavior.

### 4. Distribution Shape
- KDE plots confirm right-skewed price distributions in France and Germany.
- Spain shows a narrow, centered density, meaning low volatility.

### Key Insight
France and Germany offer high upside but come with volatility. Spain and Belgium are stable and may suit lower-risk strategies.
